## 1.3. Assess Categorical Data

```python
# load sheet Demographic and show top head
demo = pd.read_excel('KPMG.xlsx','CustomerDemographic')
demo.head()
```

```python
# print to table's info
demo.info()
```

```python
# describe only string columns (object)
demo.describe(include=[object])
```

```python
# Remove all NA row
demo_dropna = demo.dropna()
```

```python
# print all the unique value of 'gender' column 
print(demo_dropna.gender.unique())
```

```python
# Standardize the Gender column for 2 groups Male and Female, Unisex and print out a new unique value
demo_dropna = demo_dropna.replace({'gender':{'F': 'Female', 'Femal': 'Female'}})
demo_dropna.gender
                                  
```

```python
# Calculate Customer age and save to "Age" column 
demo_dropna['DOB'] = pd.to_datetime(demo_dropna.DOB)
demo['dob_y'] = demo_dropna['DOB'].dt.year
demo_dropna['age'] = (dt.datetime.now().year - demo.dob_y)
print(demo_dropna.age)


        
```

```python
# print min age, max age
print(demo_dropna.age.max())
print(demo_dropna.age.min())
```

```python
# group customer under 35 to "Young", 36 - 55 to "Middle", over 55 to "Older"
# save the result to "Age_Group" 
demo_dropna['age_group'] = pd.cut(x=demo_dropna['age'], bins=[20,35,55,91], labels=['Young', 'Middle', 'Older'])
demo_dropna
```

# 3. Data Aggregation

```python
# Calculate total customer
num_unique_customer = df_dropna_all['customer_id'].nunique()
num_unique_customer
```

```python
# Calculate number of customer by month
df_dropna_all['year_month'] = df_dropna_all['transaction_date'].dt.strftime('%Y-%m')
no_customer_per_month = df_dropna_all.groupby('year_month')['customer_id'].nunique()
no_customer_per_month
```

```python
# Calculate gross magin by month. Gross magin = list_price - standard_cost
df_dropna_all['Gross_Margin'] = df_dropna_all.list_price - df_dropna_all.standard_cost
gross_margin_per_month = df_dropna_all.groupby('year_month')['Gross_Margin'].agg(sum)
gross_margin_per_month
```

```python
# Calculate number of order and gross margin group by online_order và order_status 
df_dropna_all.groupby(['online_order', 'order_status']).agg({'list_price': ['sum', 'count']})
```

# 4. Data Visualization

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Line chart: Gross Margin By Month In 2017

#sns.relplot(x='year_month', y='Gross_Margin', data=gross_margin_per_month, kind='line')
gm_by_month = df_dropna_all[df_dropna_all['order_status'] == 'Approved'].groupby('year_month', as_index=False)['Gross_Margin'].sum()   
gm_by_month
```

```python
l = sns.relplot(x='year_month', y='Gross_Margin', data=gm_by_month, kind='line')
l.fig.set_figwidth(20.90)
plt.xlabel('Year and Month')
plt.ylabel('Gross Margin')
plt.title('Gross Margin By Month In 2017')
```

- In 2017, gross margin had 3 sharp drops in February, June, September and December
- In the remaining months, gross margin tends to increase
- Gross margin bottomed in June and peaked in October

```python
# Number of unique product and unique brand

print(df_dropna_all['product_id'].nunique())
print(df_dropna_all['brand'].nunique())
```

```python
# data for bar chart
column_data = df_dropna_all.groupby(['brand'], as_index=False)['transaction_id'].count().sort_values(by='transaction_id', ascending=False)
print(column_data)
```

```python
#create bar chart

b = sns.catplot(y='brand', x='transaction_id', data=column_data, kind='bar',palette='twilight')
b.fig.set_figwidth(10.90)
plt.ylabel("Brand")
plt.xlabel("Number of Tranctions")
plt.title("Number of Tranctions by each Brand")
```

- Most brands have the similar transactions.
- Solex has the highest volume of transactions.

```python
# Bar chart for GM by customer job_industry_category

gm = SD.groupby(['job_industry_category'], as_index=False)['list_price'].sum().sort_values('list_price',ascending=False)
gm
```

```python
# Bar chart 
b2 = sns.catplot(x='job_industry_category', y ='list_price', data=gm, kind='bar', palette='plasma')
b2.fig.set_figwidth(15, 6)
plt.xlabel('Job industry')
plt.ylabel('Gross Margin')
plt.title('Gross Margin by Job industry')
plt.show()

```

- Top 3 customer job industries that has the most gross margin is: Manufacturing, Financial Services and Health.
